```python
import librosa 
import librosa.display
import matplotlib.pyplot as plt
import os
```

```python
import os
import pandas as pd

# Initialize list to store file info
data = []

# Traverse through directories and get all .wav files
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename.endswith('.wav'):
            filepath = os.path.join(dirname, filename)
            try:
                # Extract emotion label from filename (assumes format like: 788_SAD.wav)
                emotion = filename.split('_')[1].replace('.wav', '')
                data.append({'filepath': filepath, 'emotion': emotion})
            except IndexError:
                print(f"Could not extract emotion from {filename}")

# Create DataFrame
df = pd.DataFrame(data)

# Show sample
print(df.head())
```

```python
audio_path='/kaggle/input/mtl-782-assignment-2-track-2-emotion-detection/data/data/788_SAD.wav'
```

```python
audio_signal,sr=librosa.load(audio_path,sr=None)

plt.figure(figsize=(10,4))
librosa.display.waveshow(audio_signal,sr=sr)
plt.title('waveform of 788_SAD.wav')
plt.show()
```

```python
stft=librosa.stft(audio_signal)
spectogram_db=librosa.amplitude_to_db(abs(stft),ref=np.max)
plt.figure(figsize=(10,4))
librosa.display.specshow(spectogram_db,sr=sr,x_axis='time',y_axis='log',cmap='virdis')
plt.colorbar(format='%+2.0f dB')
plt.title('spectogram')
plt.show()
```
